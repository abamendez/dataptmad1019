#### Using Seaborn plot the correlation matrix of various features.

```python
#your code here

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1, 1, 1)

sns.heatmap(titanic.corr(method='spearman'), annot=True, fmt='.2f', ax=ax);
```
